    <p style="color: #2c3e50; font-size: 135%; border-left: 15px solid #3498db; padding-left: 20px; letter-spacing: 1px;"><strong>What is a Landfill?</strong></p>
    <p>A landfill is a site used for the disposal of waste materials by burial. It is the oldest and most common form of waste treatment. Landfills are designed to safely manage and contain waste, minimizing environmental impacts.</p>
    <p><strong>How Does It Work?</strong></p>
    <p>Landfills are carefully engineered and managed. Waste is deposited in layers and compacted. A liner system is used to prevent leachate (contaminated liquid) from contaminating the surrounding environment. Gas collection systems capture methane, a byproduct of decomposition, for potential energy use.</p>
    <p><strong>What Are the Benefits?</strong></p>
    <p>Landfills provide a controlled environment for waste disposal, protecting public health and the environment. They can be sources of renewable energy through methane collection and contribute to recycling efforts through materials recovery.</p>
    <p><strong>What Is Produced?</strong></p>
    <p>Landfills produce biogas, primarily methane, which can be captured and used as a renewable energy source. Leachate, a liquid byproduct, is also managed and treated.</p>
    <p><strong>What Is the Purpose?</strong></p>
    <p>The primary purpose of landfills is to isolate waste from the surrounding environment, preventing pollution and protecting public health. They also contribute to waste-to-energy initiatives and resource recovery.</p>
    <p><strong>Environmental Benefits?</strong></p>
    <p>Properly managed landfills minimize negative environmental impacts. Methane capture reduces greenhouse gas emissions, while leachate management protects water quality. Modern landfills often incorporate habitat restoration and green space development.</p>
    <p style="font-size: 130%; color: #8e44ad; text-shadow: 2px 2px 5px rgba(0,0,0,0.2);"><strong>Landfills in the United States</strong></p>

Various technologies and practices are employed to maximize environmental protection and economic benefits.</p>

  <p>This dataset contains detailed information about various landfills and projects in the United States. Each row represents a specific landfill and project. The information in the dataset includes:</p>

  <ol>
    <li><strong>GHGRP ID:</strong> The Identification Number of the landfill as per the Greenhouse Gas Reporting Program (GHGRP).</li>
    <li><strong>Landfill ID:</strong> The unique identification number of the landfill.</li>
    <li><strong>Landfill Name:</strong> The name of the landfill.</li>
    <li><strong>State:</strong> The state in which the landfill is located.</li>
    <li><strong>City:</strong> The city in which the landfill is located or near.</li>
    <li><strong>County:</strong> The county in which the landfill is located.</li>
    <li><strong>Ownership Type:</strong> Indicates if the landfill is publicly owned, privately owned, or co-owned by public and private entities.</li>
    <li><strong>Landfill Owner Organization(s):</strong> The organization that owns the landfill.</li>
    <li><strong>Waste in Place (tons):</strong> The waste-in-place at the landfill, measured in short tons.</li>
    <li><strong>Waste in Place Year:</strong> The year corresponding to the waste-in-place at the landfill.</li>
    <li><strong>LFG Collection System In Place?</strong> Indicates whether there is a landfill gas collection system in place.</li>
    <li><strong>LFG Collected (mmscfd):</strong> The amount of landfill gas being collected, measured in million standard cubic feet per day.</li>
    <li><strong>LFG Flared (mmscfd):</strong> The amount of landfill gas flared.</li>
    <li><strong>Project ID:</strong> The unique identifying number combination for "parent" project and "expansion" number.</li>
    <li><strong>Current Project Status:</strong> The current project status.</li>
    <li><strong>Project Name:</strong> The specific project name (if applicable).</li>
    <li><strong>Project Start Date:</strong> The date the project or expansion became operational.</li>
    <li><strong>Project Finish Date:</strong> The date the project or expansion shut down.</li>
    <li><strong>Project Type Category:</strong> The category of LFG energy project type.</li>
    <li><strong>LFG Energy Project Type:</strong> The specific type of LFG energy project technology.</li>
    <li><strong>RNG Delivery Method:</strong> The method for delivery of the renewable natural gas.</li>
    <li><strong>LFG Use Details:</strong> Details about the use of landfill gas.</li>
    <li><strong>Actual MW Generation:</strong> The actual or estimated electric generation in megawatts for electricity-generating projects.</li>
    <li><strong>Rated MW Capacity:</strong> The rated capacity in megawatts for electricity-generating projects.</li>
    <li><strong>LFG Flow to Project (mmscfd):</strong> The amount of landfill gas flowing to the project or that will flow to the project when it becomes operational.</li>
    <li><strong>End User(s):</strong> Information about the end users of the project.</li>
    <li><strong>Project Developer(s):</strong> Information about the developers of the project.</li>
    <li><strong>Project Owner(s):</strong> Information about the owners of the project.</li>
    <li><strong>Product Supplier(s):</strong> Information about the suppliers of the project.</li>
    <li><strong>Self-Developed:</strong> Information about whether the project was self-developed.</li>
    <li><strong>Current Year Emission Reductions (MMTCO2e/yr) - Direct:</strong> The direct methane reductions by the energy project for the current year.</li>
    <li><strong>Current Year Emission Reductions (MMTCO2e/yr) - Avoided:</strong> The avoided carbon dioxide emission reductions by the energy project for the current year.</li>

```python
# Recreate the analysis at the state level
state_analysis = df.groupby('State').agg(
    total_Landfills=('Landfill Name', 'count'),
    public_Landfills=('Ownership Type', lambda x: (x == 'Public').sum()),
    private_Landfills=('Ownership Type', lambda x: (x == 'Private').sum())
).reset_index()

# Get the top 10 states with the most facilities
top_10_states = state_analysis[['State', 'total_Landfills']].nlargest(10, 'total_Landfills')

# Calculate the number of facilities by ownership type in the top 10 states
ownership_by_top_10_states = df[df['State'].isin(top_10_states['State'])]
ownership_by_top_10_states = ownership_by_top_10_states.groupby(['State', 'Ownership Type']).size().reset_index(name='count')

# Create a pivot table to make the data suitable for plotting
ownership_state_top_10_pivot = ownership_by_top_10_states.pivot_table(index='State', columns='Ownership Type', values='count', fill_value=0)

# Sort
ownership_state_top_10_pivot = ownership_state_top_10_pivot.loc[top_10_states['State']]

# Plot the graph
ownership_state_top_10_pivot.plot(kind='bar', stacked=True, figsize=(12,6), color=['blue', 'red'])
plt.title('Top 10 States by Number of Landfills (By Ownership Type)')
plt.xlabel('State')
plt.ylabel('Number of Landfills')
plt.xticks(rotation=45)
plt.legend(title='Ownership Type', labels=['Public', 'Private'])
plt.tight_layout()
plt.savefig('Top 10 States by Number of Landfills (By Ownership Type).png')
plt.show()

```

```python
# Grouping by state for analysis
state_analysis = df.groupby('State').agg(
    total_facilities=pd.NamedAgg(column='Landfill Name', aggfunc='count'),
    public_facilities=pd.NamedAgg(column='Ownership Type', aggfunc=lambda x: (x == 'Public').sum()),
    private_facilities=pd.NamedAgg(column='Ownership Type', aggfunc=lambda x: (x == 'Private').sum()),
    renewable_facilities=pd.NamedAgg(column='RNG Delivery Method', aggfunc=lambda x: (x == 'Yes').sum()),
    total_waste=pd.NamedAgg(column='Waste in Place (tons)', aggfunc='sum')
).reset_index()

state_analysis = state_analysis.sort_values(ascending=True, by="renewable_facilities")

# Visualize the table
fig, ax = plt.subplots(figsize=(12, 8)) 
ax.axis('tight')
ax.axis('off')
the_table = ax.table(cellText=state_analysis.values, colLabels=state_analysis.columns, loc='center', cellLoc='center')

# Styling the table
the_table.auto_set_font_size(False)
the_table.set_fontsize(12)
the_table.scale(1.2, 2) 
for (i, j), cell in the_table.get_celld().items():
    cell.set_edgecolor('black')
    if i == 0: 
        cell.set_text_props(fontweight='bold', color='white')
        cell.set_facecolor('darkblue')
    elif i % 2 == 0: 
        cell.set_facecolor('lightgray')

# Save the table
plt.savefig('state_analysis.png', dpi=500, bbox_inches='tight')
```
